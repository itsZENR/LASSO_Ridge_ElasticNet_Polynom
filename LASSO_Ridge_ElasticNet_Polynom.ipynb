```python
#@title Библиотеки
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url = '/content/diabetes.csv'
df = pd.read_csv(url)
df
```

```python
X = df[['SkinThickness', 'BMI']]
X
```

```python
y = df[['Insulin']]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25,random_state=1)

print(np.shape(X_train))
print(np.shape(y_train))

print(np.shape(X_test))
print(np.shape(y_test))
```

```python
# LASSO регрессия
from sklearn.linear_model import Lasso
lasso_regression = Lasso(alpha=0.1) # alpha — величина регуляризации

# обучение
lasso_regression.fit(X_train, y_train)

# вывод точности предсказания
print(lasso_regression.score(X_test, y_test))
print(lasso_regression.score(X_train, y_train))
```

```python
# Ridge регрессия
from sklearn.linear_model import Ridge
ridge_regression = Ridge(alpha=0.1) # alpha — величина регуляризации

# обучение
ridge_regression.fit(X_train, y_train)

# вывод точности предсказания
print(ridge_regression.score(X_test, y_test))
print(ridge_regression.score(X_train, y_train))
```

```python
# ElasticNet регрессия
from sklearn.linear_model import ElasticNet

elasticnet_regression = ElasticNet(alpha=0.1) # alpha — величина регуляризации

# обучение
elasticnet_regression.fit(X_train, y_train)

# вывод точности предсказания
print(elasticnet_regression.score(X_test, y_test))
print(elasticnet_regression.score(X_train, y_train))
```

```python
# Polynom регрессия
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LinearRegression

for i in [2, 5, 10]:
  poly_regression= PolynomialFeatures(degree= i)  
  x_poly= poly_regression.fit_transform(X_train)  

  lin_reg =LinearRegression()  
  lin_reg.fit(x_poly, y_train)  

  print(lin_reg.score(x_poly, y_train))
```
